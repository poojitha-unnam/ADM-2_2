```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
          "Insulin",
          "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
data = pd.read_csv(url, names=columns)

# Set visual style
sns.set(style="whitegrid", palette="pastel")
plt.figure(figsize=(12, 6))

# 1. Count plot of Outcome (Diabetes or Not)
plt.subplot(1, 2, 1)
sns.countplot(data=data, x="Outcome")
plt.title("Diabetes Count (0 = No, 1 = Yes)")

# 2. Histogram of all numerical features
data.hist(bins=20, figsize=(14, 10), color='skyblue')
plt.suptitle("Distribution of Features", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])

# 3. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")

# 4. Box plots to visualize outliers
plt.figure(figsize=(15, 10))
for i, column in enumerate(data.columns[:-1]):
    plt.subplot(3, 3, i+1)
    sns.boxplot(data=data, y=column, x='Outcome')
    plt.title(f'{column} vs Outcome')
plt.tight_layout()

# 5. Pairplot (can be slow with large datasets)
# Uncomment the line below if you want to see relationships between features
# sns.pairplot(data, hue="Outcome", corner=True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://archive.ics.uci.edu/static/public/17/data.csv"
columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
          "Insulin",
          "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
data = pd.read_csv(url, names=columns)

# Set style
sns.set(style="whitegrid")

# 1. Pairplot for first few features (colored by Outcome)

plt.suptitle("Pairplot of selected features", y=1.02)
plt.show()

# 2. Heatmap for correlation
plt.figure(figsize=(10, 8))


plt.title("Correlation Heatmap")
plt.show()

# 3. Boxplots for each feature vs Outcome
features = data.columns[:-1]
plt.figure(figsize=(16, 10))
for idx, feature in enumerate(features):
    plt.subplot(3, 3, idx + 1)
    sns.boxplot(x="Outcome", y=feature, data=data, palette="pastel")
    plt.title(f"{feature} vs Outcome")
plt.tight_layout()
plt.show()

# 4. Distribution plots (Histograms)
plt.figure(figsize=(16, 10))
for idx, feature in enumerate(features):
    plt.subplot(3, 3, idx + 1)
    sns.histplot(data[feature], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()

# 5. Countplot of Outcome
sns.countplot(x='Outcome', data=data, palette='Set1')
plt.title("Outcome Count")
plt.xticks([0, 1], ['Non-diabetic', 'Diabetic'])
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Kidney dataset
url = "https://archive.ics.uci.edu/static/public/336/data.csv"
columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
          "Insulin",
          "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
data = pd.read_csv(url, names=columns)

# Set visualization style
sns.set(style="whitegrid")

# 1. Pairplot of selected features

plt.suptitle("Pairplot of selected features", y=1.02)
plt.show()

# 2. Correlation heatmap
plt.figure(figsize=(10, 8))


plt.title("Correlation Heatmap")
plt.show()

# 3. Boxplots: Feature vs Outcome
features = data.columns[:-1]
plt.figure(figsize=(16, 10))
for idx, feature in enumerate(features):
    plt.subplot(3, 3, idx + 1)
    sns.boxplot(x="Outcome", y=feature, data=data, palette="pastel")
    plt.title(f"{feature} vs Outcome")
plt.tight_layout()
plt.show()

# 4. Histograms for feature distributions
plt.figure(figsize=(16, 10))
for idx, feature in enumerate(features):
    plt.subplot(3, 3, idx + 1)
    sns.histplot(data[feature], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()

# 5. Countplot for class distribution
sns.countplot(x='Outcome', data=data, palette='Set1')
plt.title("Outcome Count")
plt.xticks([0, 1], ['Non-Diseased', 'Diseased'])
plt.show()
```
